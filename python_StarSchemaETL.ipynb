```python
df['JobDate']= pd.to_datetime(df['JobDate'])
df['year']= df['JobDate'].dt.year
df['MonthYr']= df['JobDate'].dt.to_period('M')
df['QuarterYr'] = pd.PeriodIndex(df.JobDate, freq='Q')
```

```python
df.head()
```

```python
a= df['Employer'].value_counts().reset_index()
a.reset_index(inplace=True)
a.columns=['EmployerID', 'Employer', 'counts']
a
```

```python
df= pd.merge(df, a[['Employer', 'EmployerID']])
```

```python
df['temp']= df['Lat'].astype('str')+df['Lon'].astype('str')+df['City']+df['County']
df['temp']
```

```python
import hashlib
# df['survey_id'] = (df.employee_id + df.customer_id + df.timestamp).apply(
#     lambda s: hashlib.md5(s).digest().encode('base64')[: 10])

df['LocID']= df['temp'].apply(lambda x: hashlib.md5(x.encode('utf-8')).hexdigest())
```

```python
df.drop(columns=['NAICS3', 'temp'], inplace=True)
df.shape
```

```python
df['LocID'].value_counts()
```

```python
dfl= df[['LocID', 'City', 'State', 'County', 'Lat', 'Lon']]
dfl.drop_duplicates(inplace=True)
dfl.reset_index(inplace=True, drop=True)
dfl.shape
```

```python
dfl['LocID'].value_counts()
```

```python
df= df[df['LocID']!='c9d4de8cd171f87b15bb98e968421aaf']
df= df[df['LocID']!='d4793d01e73ff9e0c0ce242093e9c435']
df.reset_index(inplace=True, drop=True)
df
```

# DIMENSION CREATION

```python
# dim_occupation
dfo= df[['OccFam', 'OccFamName']]
dfo.drop_duplicates(inplace=True)
dfo.reset_index(inplace=True, drop=True)
dfo.shape
```

```python
# dim_soc
dfs= df[['SOC', 'SOCName']]
dfs.drop_duplicates(inplace=True)
dfs.reset_index(inplace=True, drop=True)
dfs.shape
```

```python
# dim_onet
dfon= df[['ONET', 'ONETName']]
dfon.drop_duplicates(inplace=True)
dfon.reset_index(inplace=True, drop=True)
dfon.shape
```

```python
# dim_emp
dfe= df[['EmployerID', 'Employer']]
dfe.drop_duplicates(inplace=True)
dfe.reset_index(inplace=True, drop=True)
dfe.shape
```

```python
# dim_sector
dfse= df[['Sector', 'SectorName']]
dfse.drop_duplicates(inplace=True)
dfse.reset_index(inplace=True, drop=True)
dfse.shape
```

```python
# dim_location
dfl= df[['LocID', 'City', 'State', 'County', 'Lat', 'Lon']]
dfl.drop_duplicates(inplace=True)
dfl.reset_index(inplace=True, drop=True)
dfl.shape
```

```python
# dim_time
dft= df[['JobDate', 'year', 'MonthYr', 'QuarterYr']]
dft.drop_duplicates(inplace=True)
dft.reset_index(inplace=True, drop=True)
dft.shape
```

```python
# dim_role
dfr= df[['JobID', 'CleanTitle', 'CanonTitle']]
dfr.drop_duplicates(inplace=True)
dfr.reset_index(inplace=True, drop=True)
dfr.shape
```
